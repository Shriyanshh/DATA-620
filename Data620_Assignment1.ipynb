```python
# Import libraries
import networkx as nx
import matplotlib.pyplot as plt
import networkx.generators.small as ns
```

```python
# Plot the Krackhardt kite graph
g = ns.krackhardt_kite_graph()
nx.draw(g, with_labels=True)
plt.title("Krackhardt Kite Graph")
plt.show()
```

```python
# Let's relabel the nodes with the actual names
mapping = {
    0: "Beverly", 1: "Andre", 2: "Ed", 3: "Diane", 4: "Carol",
    5: "Garth", 6: "Fernando", 7: "Heather", 8: "Ike", 9: "Jane"
}
# Plot the Krackhardt kite graph with named nodes
G = nx.relabel_nodes(g, mapping)
nx.draw(G, with_labels=True)
plt.title("Krackhardt Kite Graph with Named Nodes")
plt.show()
```
